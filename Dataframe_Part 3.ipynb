```python
#No Sorting
```

```python
mtcars.groupby('model',sort=False).first()
```

```python
#Grouping two columns
```

```python
mtcars.groupby(['model','mpg']).first()
```

```python
#.agg
```

```python
mtcars.groupby(['model','mpg']).agg(np.mean)
```

```python
#multiple operations 
```

```python
mtcars.groupby(['model','mpg']).agg([min, max])
```

```python
#.agg on multiple columns
```

```python
mtcars.groupby(
    ['model','mpg']
).agg(
    {
        'mpg': [min, max],
        'carb': "count",
        'gear': [min, max]
    }
)
```

```python
#.apply()
```

```python
mtcars.groupby('model').apply(lambda grp: grp.count())
```

```python
#Transform
```

```python
mtcars["trans"] = mtcars.groupby('model')["mpg"].transform('mean')
```
